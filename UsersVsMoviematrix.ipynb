<h4><b>-> Building SVD from scratch</b></h4>

```python
def power_iteration(A, num_iterations=300, tol=1e-6):
    n = A.shape[0]
    eigenvalues = []
    eigenvectors = []
    
    for _ in range(num_iterations):
        b = np.random.rand(n)  # Initialize a random vector
        for _ in range(num_iterations):
            b_next = np.dot(A, b)
            b_next_norm = np.linalg.norm(b_next, ord=np.inf)
            if b_next_norm == 0:
                b_next_norm = 1
            b_next = b_next / b_next_norm
            if np.linalg.norm(b - b_next, ord=2) < tol:
                break
            b = b_next
        eigenvalue = np.dot(b, np.dot(A, b))  # Rayleigh quotient for the eigenvalue
        eigenvalues.append(eigenvalue)
        eigenvectors.append(b)
        
        # Deflate the matrix
        A = A - eigenvalue * np.outer(b, b)
        
    return np.array(eigenvalues), np.array(eigenvectors).T
```

```python
def svd(R):
    # Step 1: Compute the covariance matrix
    A = np.dot(R.T, R)

    # Step 2: Compute the eigenvalues and eigenvectors of the covariance matrix
#     eigenvalues, V = np.linalg.eig(A)
    eigenvalues, V = power_iteration(A)

    # Step 3: Compute the singular values and right singular vectors
    sigma = np.sqrt(np.abs(eigenvalues)) * np.sign(eigenvalues)
    Vt = V.T

    # Step 4: Compute the left singular vectors from the right singular vectors
    U = np.dot(R, V) / sigma

    return U, sigma, Vt

def svd_new(R, k):
    
    mask = ~np.isfinite(R).any(axis=1)
    if np.sum(mask) > 0:
        print('Non-finite values detected in R. Removing rows...')
        R = R[~mask, :]
        
    U, sigma, Vt = svd(R)
    
    # Reshape U and Vt if necessary
    if U.ndim == 1:
        U = U.reshape(-1, 1)
    if Vt.ndim == 1:
        Vt = Vt.reshape(1, -1)
    
    U = U[:, :k]
    sigma = np.diag(sigma[:k])
    Vt = Vt[:k, :]
    return U, sigma, Vt
```

<h4><b>-> Here we will be getting User vector, Diagonalization Matrix, Movie vector</b></h4>
<h4><b>-> Sigma's with very small weight might be noise. So just remove them </b></h4>
<h4><b>-> Each rating is obtained as a weighted dot product between corresponding user vector and item vector with singular values as the weight</b></h4>

<h2><b>Using k means</b></h2>


<h4><b>-> Building k means from scratch</b></h4>

```python


class KMeans:
    def _init_(self):
        pass

    def fit(self, X, n_clusters=8, max_iter=10,random_state=None):
        self.n_clusters = n_clusters
        self.max_iter = max_iter
        self.random_state = random_state
        if isinstance(X, pd.DataFrame):
            X = X.values

        if self.random_state is not None:
            np.random.seed(self.random_state)
            
        
        # Initialize centroids randomly
        centroids = X[np.random.choice(X.shape[0], self.n_clusters, replace=False)]
        
        for _ in range(self.max_iter):
            # Compute distances between each point and centroids
            distances = np.sqrt(((X[:, np.newaxis] - centroids) ** 2).sum(axis=2))
            
            # Assign each data point to the closest centroid
            labels = np.argmin(distances, axis=1)
            
            # Update centroids
            new_centroids = np.array([X[labels == k].mean(axis=0) for k in range(self.n_clusters)])
            
            # Check for convergence
            if np.all(centroids == new_centroids):
                break
                
            centroids = new_centroids
        
        self.labels_ = labels
        self.cluster_centers_ = centroids

    def predict(self, X):
        if isinstance(X, pd.DataFrame):
            X = X.values
        distances = np.sqrt(((X[:, np.newaxis] - self.cluster_centers_) ** 2).sum(axis=2))
        return np.argmin(distances, axis=1)
```

<h4><b>-> We will call the Kmeans function</b></h4>


```python

from sklearn.metrics import pairwise_distances_argmin_min

temp = all_user_predicted_ratings.reset_index(drop=True, inplace=True)

# Apply KMeans clustering on the predicted ratings
kmeans = KMeans()
kmeans.fit(all_user_predicted_ratings, n_clusters=120, random_state=42)
```

<h4><b>Distance(square root) based recommendation[kmeans]</b></h4>
<h4><b>-> Compute normalized distances between each point and centroids</b></h4>
<h4><b>-> Getting Recommendations based on that</b></h4>

```python
# Function to recommend movies for a particular user
def recommend_movies_for_user(user_id, k=30):
    # Find the index of the user
    user_index = user_id - 1  # Assuming user IDs start from 1
    # Transform user's ratings into the reduced dimensional space
    user_ratings_embedding = all_user_predicted_ratings.iloc[user_index].values.reshape(1, -1)
    # Find the cluster label of the user
   # cluster_label = kmeans.labels_[user_index]
    cluster_label = kmeans.predict(user_ratings_embedding)[0]
    # Find the users in the same cluster
    cluster_indices = np.where(kmeans.labels_ == cluster_label)[0]
    # Compute distances between the given user and all other users in the cluster
    distances = pairwise_distances_argmin_min(user_ratings_embedding, all_user_predicted_ratings.iloc[cluster_indices])[1]
    # Sort by distance and get top k users
    recommended_user_indices = [cluster_indices[i] for i in np.argsort(distances)]
    # Get movie recommendations from the top user in the cluster
    recommended_movies = all_user_predicted_ratings.iloc[recommended_user_indices[0]].sort_values(ascending=False).index[:k]
    return recommended_movies

# Example usage
user_id = 0  # Input your user ID
recommendations = recommend_movies_for_user(user_id)
print("Recommendations for User {}: ".format(user_id))
for i, movie_id in enumerate(recommendations, 1):
    movie_name=df_movies[df_movies['MovieID']==movie_id]['Title']
    movie_genres=df_movies[df_movies['MovieID']==movie_id]['Genres']
    print("{}. Movie Name: {} Movie_Genres:{} ".format(i, movie_name.values,movie_genres.values))
```

```python
# Function to recommend movies for a particular user
def recommend_movies_for_user(user_id, k=30):
    # Find the index of the user
    user_index = user_id - 1  # Assuming user IDs start from 1
    # Transform user's ratings into the reduced dimensional space
    user_ratings_embedding = all_user_predicted_ratings.iloc[user_index].values.reshape(1, -1)
    # Find the cluster label of the user
    cluster_label = kmeans.labels_[user_index]
    # Find the users in the same cluster
    cluster_indices = np.where(kmeans.labels_ == cluster_label)[0]
    return cluster_indices

# Example usage
user_id = 0  # Input your user ID
cluster_user_ids = recommend_movies_for_user(user_id)
cluster_user_ids
```

```python
def pearson_correlation(user_ratings, other_user_ratings):
    # Mean of user ratings
    mean_user = np.mean(user_ratings)
    mean_other_user = np.mean(other_user_ratings)
    
    # Compute covariance
    cov = np.sum((user_ratings - mean_user) * (other_user_ratings - mean_other_user))
    
    # Compute standard deviations
    std_user = np.sqrt(np.sum((user_ratings - mean_user) ** 2))
    std_other_user = np.sqrt(np.sum((other_user_ratings - mean_other_user) ** 2))
    
    # Compute Pearson correlation coefficient
    pearson_corr = cov / (std_user * std_other_user)
    
    return pearson_corr



new_user=0
#getting the rating vector for the new user
user_ratings = all_user_predicted_ratings.iloc[new_user]
#getting the rating vector for the cluster user
cluster_user_ratings = all_user_predicted_ratings.iloc[cluster_user_ids]

# Compute Pearson correlation with other users

user_similarity = {}
for other_user_id, other_user_ratings in cluster_user_ratings.iterrows():
    if other_user_id != new_user:
        similarity = pearson_correlation(user_ratings, other_user_ratings)
        user_similarity[other_user_id] = similarity
        
user_similarity = pd.Series(user_similarity)
# Take a look at the data
user_similarity.head()


# Number of similar users
n = 10
# Get top n similar users
similar_users = user_similarity.sort_values(ascending=False)[:n]
# Print out top n similar users
print(f'The similar users for user {user_id} are: \n',similar_users)
```

<h2><b>Calculating similarity score</b></h2>

```python
# A dictionary to store movie scores
item_score = {}

# Loop through the similar movies
for i in cluster_user_ratings.columns: #movie vector
    
  # Retrieve the ratings for movie i
  movie_rating = cluster_user_ratings[i]
    
  # Create a variable to store the movie score
  total = 0
    
  # Create a variable to store the number of scores
  count = 0
    
  # Loop through similar users
  for u in similar_users.index:
        
    # Score is the sum of user similarity score multiply by the movie rating
    score = similar_users[u] * movie_rating[u]
    
    # Add the score to the total score for the movie so far
    total += score
    count +=1
    
  # Get the average score for the item
  item_score[i] = total / count # Formula: Summation(user_similarity_value * rating given by the user)/total number of similar users.
    
# Convert dictionary to pandas dataframe
item_score = pd.DataFrame(item_score.items(), columns=['movie', 'movie_score'])
    
# Sort the movies by score
top_recommended_movies = item_score.sort_values(by='movie_score', ascending=False)

top_recommended_movies.rename(columns={'movie': 'MovieID'}, inplace=True)



# Select top k movies
k = 20
top_recommended_movies['MovieID'] = top_recommended_movies['MovieID'].astype(int)
top_recommended_movies
```
